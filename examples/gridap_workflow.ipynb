```julia
import Gridap
using Gridap.Geometry
using Gridap.FESpaces
using Gridap.ReferenceFEs
using Gridap.Arrays
using Gridap.Fields
using Gridap.CellData
using FillArrays
```

```julia
domain = (0.0, 1.0, 0.0, 1.0)
partition = (2,3)
model = CartesianDiscreteModel(domain,partition)
pol = Polytope(Fill(HEX_AXIS,2)...)
reffe = LagrangianRefFE(Float64,pol,2)

f(x) = 1.0
g(x) = 0.0

Vₕ = TestFESpace(model,reffe;conformity=:H1,dirichlet_tags=["boundary"])
Uₕ = TrialFESpace(Vₕ, g)
Ω = Triangulation(model)
dΩ = Measure(Ω,2)

# Weak Form
a(u,v) = ∫( ∇(v)⋅∇(u) )*dΩ
a(u,v) = ∇(v)⋅∇(u)
b(v) = ∫( v*f )*dΩ 

op = AffineFEOperator(a,b,Uₕ,Vₕ)
```
